```python
from collections import Counter
class Preprocessing():
    def __init__(self,  table_name):
        
        # on which table are we working on
        self.table_name = table_name
    
    # change the data format from e.g. 10 May, 2020 to 2020-05-10
    def change_time(self, column_name):
        self.table_name[column_name]=[ datetime.datetime.strptime(self.table_name[column_name][i], '%B %d, %Y')\
                                      for i in range(0,len(self.table_name[column_name]))]
    
    # find the most common values in a column
    def most_common(self, column_name):
        return Counter(self.table_name[column_name]).most_common(5)
    
    # find the most common unigrams in titles
    def most_common_unigrams_in_titles(self):
        titles=""
        for i in range(0, self.table_name.shape[0]):#press.shape[0]):
            titles = titles+" " + self.table_name['title'][i]

        tokenized_text = word_tokenize(titles.lower())
        nltk_text = nltk.Text(tokenized_text)
        titles_unigrams = nltk.FreqDist(nltk_text).most_common()
        return titles_unigrams
    
    # compute sentiment for a column
    def count_sentiment(self, column_name):
        analyzer_speech = SentimentIntensityAnalyzer()
        text_speech = self.table_name[column_name]
        scores_speech = []
        for txt in text_speech:
            speech_analysis = analyzer_speech.polarity_scores(txt)
            scores_speech.append(speech_analysis)
            
        data_new = pd.DataFrame(scores_speech)
        data_new=data_new.rename(columns = {'pos':'pos_{}'.format(column_name),\
                                  'compound':'compound_{}'.format(column_name), \
                                   'neg':'neg_{}'.format(column_name), \
                                   'neu':'neu_{}'.format(column_name)})
        self.table_name= pd.concat([self.table_name,data_new], axis=1)
        return self.table_name
```

```python
import nltk
from nltk.corpus import reuters
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer 
import re
import string
```

```python
# create a class that computes text features

class TextFeatures:

    def __init__(self,  table_name):
        
        self.table_name = table_name
    
    #counts number of sentences and words in the speech
    def _count_sentences(self):
        number_sentences = []
        number_words = []
        for i in range (self.table_name.shape[0]):
            sentences = nltk.sent_tokenize(self.table_name.iat[i,6].replace('\n',' ').strip())
            default_tokens = word_tokenize(self.table_name.iat[i,6])
            number_sentences.append(len(sentences))
            number_words.append(len(default_tokens))

        number_sentences = pd.Series(number_sentences)
        number_sentences.name = 'number of sentences'
        number_words = pd.Series(number_words)
        number_words.name = 'number of words'
        
        return number_sentences, number_words
    
    def _count_categories(self):
        regex_nouns = re.compile('^N.*')
        regex_verbs = re.compile('^V.*')
        regex_adjectives = re.compile('^J.*')
        regex_adverbs = re.compile('^R.*')

        #compute the percentage of nouns, verbs, adjectives and adverbs
        perc_nouns = []
        perc_verbs = []
        perc_adjectives = []
        perc_adverbs = []

        for i in range (self.table_name.shape[0]):
            default_tokens = word_tokenize(self.table_name.iat[i,6])
            default_pos = nltk.pos_tag(default_tokens)

            nouns = 0
            verbs = 0
            adjectives = 0
            adverbs = 0

        #count the number nouns, verbs, adjectives and adverbs in the text
            for l in default_pos:
                if regex_nouns.match(l[1]):
                    nouns = nouns +1
                elif regex_verbs.match(l[1]):
                    verbs = verbs + 1
                elif regex_adjectives.match(l[1]):
                    adjectives = adjectives +1
                elif regex_adverbs.match(l[1]):
                    adverbs = adverbs + 1

            perc_nouns.append(nouns/len(default_tokens))
            perc_verbs.append(verbs/len(default_tokens))
            perc_adjectives.append(adjectives/len(default_tokens))
            perc_adverbs.append(adverbs/len(default_tokens))

        percentage_nouns = pd.Series(perc_nouns)
        percentage_nouns.name = 'percentage of nouns'
        percentage_verbs = pd.Series(perc_verbs)
        percentage_verbs.name = 'percentage of verbs'
        percentage_adjectives = pd.Series(perc_adjectives)
        percentage_adjectives.name = 'percentage of adjectives'
        percentage_adverbs = pd.Series(perc_adverbs)
        percentage_adverbs.name = 'percentage of adverbs'
        
        return percentage_nouns, percentage_verbs, percentage_adjectives, percentage_adverbs
    
    def _compute_ngrams(self):
        #create the punctation sequence 
        punct_seq = [c+"\"" for c in string.punctuation ]+ ["\""+c for c in string.punctuation ]
        
        #here starts the preprocessing and the after preprocessing
        most_common_unigrams = []
        most_common_bigrams = []
        most_common_trigrams = []
        all_texts = []
        for i in range (self.table_name.shape[0]):
            my_tokenized_text = word_tokenize(self.table_name.iat[i,6])
            my_nltk_text = nltk.Text(my_tokenized_text)

            #eliminate stopwords and put lower case
            my_text_lowercase_and_nostopwords = [w.lower() for w in my_nltk_text if w.lower() not in stopwords.words('english')]
            #eliminate punctuation
            my_text_nostringpunctuation = [w for w in my_text_lowercase_and_nostopwords if w not in string.punctuation]
            #eliminate punctuation sequence
            text = [w for w in my_text_nostringpunctuation if w not in punct_seq]
            
            # lemmatizing
            lemmatizer = WordNetLemmatizer()
            text = [lemmatizer.lemmatize(word) for word in text]
            
            freq_dist_unigrams = nltk.FreqDist(text)
            most_common_unigrams.append(freq_dist_unigrams.most_common(1)[0][0])

            text_bigrams = nltk.bigrams(text)
            freq_dist_bigrams = nltk.FreqDist(text_bigrams)
            most_common_bigrams.append(freq_dist_bigrams.most_common(1)[0][0])

            text_trigrams = nltk.trigrams(text)
            freq_dist_trigrams = nltk.FreqDist(text_trigrams)
            most_common_trigrams.append(freq_dist_trigrams.most_common(1)[0][0])

        #most frequent uni,bi and tri grams
        unigram = pd.Series(most_common_unigrams)
        unigram.name = 'most common unigram'
        bigram = pd.Series(most_common_bigrams)
        bigram.name = 'most common bigram'
        trigram = pd.Series(most_common_trigrams)
        trigram.name = 'most common trigram'
        
        return unigram, bigram, trigram
    
    # concatenate to create the final dataframe
    def add_features(self):
        number_sentences,number_words=self._count_sentences()
        percentage_nouns, percentage_verbs, percentage_adjectives, percentage_adverbs = self._count_categories()
        unigram,bigram,trigram = self._compute_ngrams()
        
        table_final = pd.concat([self.table_name,number_sentences,number_words, percentage_nouns,
                        percentage_verbs,percentage_adjectives,percentage_adverbs, unigram,bigram,trigram], axis=1)
        return table_final
```
